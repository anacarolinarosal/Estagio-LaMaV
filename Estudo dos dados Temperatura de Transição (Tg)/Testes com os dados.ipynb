Índice de refração: medido para um comprimento de onda de 589.3nm - Adimensional

Número de Abbe: mede a dispersão cromática - Adimensional

CTE: coeficiente linear de expansão térmica abaixo da temperatura de transição - Unidade: K^{-1}

Módulo de elasticidade de Young - Unidade: GPa

Tg: teperatura de transição - Unidade: K

Tliquidus: Temperatura de fusão - Unidade: K.

T0 A T12: Tn é a temperatura onde o logaritmo na base 10 da viscosidade( em Pa.s) vale n - Unidade: K.

ViscisityAt773K to ViscosityAt273K: valor do logaritmo na base 10 da viscosidade (em Pa.s) em uma determinada temperatura - Adimensional.
